`thinkdsp` is a module that accompanies _Think DSP_ and provides classes and functions for working with signals.

[Documentation of the thinkdsp module is here](http://greenteapress.com/thinkdsp.html). 

```python
# Get thinkdsp.py

import os

if not os.path.exists('thinkdsp.py'):
    !wget https://github.com/AllenDowney/ThinkDSP/raw/master/code/thinkdsp.py

import thinkdsp as dsp
```

```python
wave = dsp.read_wave('titanium.wav')
wave.make_audio()
```

The `ys` attribute is a NumPy array that contains the values from the signal.

```python
print('Number of samples', len(wave.ys))
print('Timestep in ms', 1 / wave.framerate * 1000)
```

```python
segment = wave.segment()
segment.plot()
dsp.decorate(xlabel='Time (s)')
```

```python
segment = wave.segment(start=0, duration=0.01)
segment.plot()
dsp.decorate(xlabel='Time (s)')
```

```python
spectrum = wave.make_spectrum()
spectrum.plot()
dsp.decorate(xlabel='Frequency (Hz)', ylabel='Amplitude')
```

```python
# original song
spectrum.make_wave().make_audio()
```

```python
# background music extraction
spectrum = wave.make_spectrum()   # FFT feature extraction
spectrum.low_pass(840)            # cut off higher frequencies
spectrum.plot()
dsp.decorate(xlabel='Frequency (Hz)', ylabel='Amplitude')
spectrum.make_wave().make_audio() # reverse FFT 
```

```python
# vocals extraction
spectrum = wave.make_spectrum()   # FFT
spectrum.high_pass(1100)          # cut off lower frequencies this time
spectrum.plot()
dsp.decorate(xlabel='Frequency (Hz)', ylabel='Amplitude')
spectrum.make_wave().make_audio() # reverse FFT
```
